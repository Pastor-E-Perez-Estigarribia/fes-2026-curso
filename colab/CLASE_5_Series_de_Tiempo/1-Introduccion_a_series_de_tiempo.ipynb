## **Casos prácticos de análisis y previsión de series temporales**

- **Finanzas:** Uno de los ejemplos más populares de previsión de series temporales es la predicción del precio de las acciones. Aunque a todos nos gustaría saber cómo calcular qué acción será el próximo Google o Amazon, este es también uno de los casos de uso más difíciles de la previsión de series temporales.

- **Meteorología:** Otro de los casos de uso más conocidos de la predicción de series temporales es la predicción meteorológica. También resulta ser un ejemplo excelente para demostrar que las previsiones con un horizonte de predicción más corto (por ejemplo, el tiempo que hará mañana) suelen ser más fiables que las previsiones con un horizonte de predicción más largo (por ejemplo, el tiempo que hará dentro de una semana).

- **Demanda:** Prácticamente todo lo que tiene demanda es un buen caso de uso para el análisis y la previsión de series temporales. Algunos ejemplos son la generación de energía, el inventario de una tienda, la dotación de personal, el número de pasajeros de un autobús o un vuelo, o el tráfico en una autopista.

- **Epidemias o ploblematicas de salud publica:** Las epidemias son dinamicas que transcurren en el tiempo. Asi mismo, algunos problemas de salud pública, como las tendencias en la mortaldiad, o exceso de mortalidad son fenomenos que pueden indexarse en el tiempo.  

La previsión del comportamiento en series temporales dependerá de:

**¿En qué medida conocemos los factores que influyen?**

**¿De cuántos datos disponemos?**

**¿Cuál es la similitud entre el futuro y el pasado?**





Una serie de datos de tiempo significa que los datos se registran en diferentes períodos o intervalos de tiempo. Los datos de series temporales pueden ser de tres tipos:

  1. **Datos de series de tiempo:** las observaciones de los valores de una variable registradas en diferentes puntos en el tiempo se denominan datos de series de tiempo.

  2. **Datos transversales (Cross sectional data):**  Son los datos de una o más variables registrados en el mismo punto en el tiempo.

  3. **Datos agrupados (Pooled data):** es la combinación de datos de series temporales y datos transversales.

## **Tipos de series de tiempo**

Podemos describir dos tipos de series temporales según su comportamiento

1. **Seríes de tiempo deterministas**

Se pueden describir por una espresión analitica. No muestra comportamiento aleatorio o probabilistico. El pasado y el futuro se pueden especificar explícitamente mediante valores dedivados en un momento.

2. **Series temporales no deterministas**

No se puede describir de forma exacta por una expresión analítica ya que tiene aspecto aleatorio. Dado un conjunto de onservaciones, se dice que una serpie es no deterministica si satisface que: la función que gobierna los datos es aleatoria.

## **Componentes de la serie de tiempo**

El comportamiento en una seríe temporal estará gobernado por los siguientes componentes:

- **Nivel:** el valor medio.

- **Tendencia:** tendencia creciente o decreciente a largo plazo.

- **Estacionalidad:** pauta que se repite a corto plazo en períodos fijos, por ejemplo, semanal, mensual o anualmente.

- **Ciclo:** pauta que se repite a corto plazo (no en periodos fijos).

- **Ruido:** fluctuaciones irregulares o incluso aleatorias a corto plazo.

Los patrones estacionales y cíclicos pueden ser difíciles de distinguir en los datos de series temporales. Si la frecuencia de un patrón recurrente es constante con respecto al calendario, entonces el patrón es estacional. En caso contrario, es cíclico.

Por simplicidada podemos usar la *Descomposición fundamental* y pensar en una serie de tiempo como compuesta de tres partes distintas. La tendencia , la estacionalidad y el ruido aleatorio.

## **Descomposición ETS:**

La descomposición **ETS** se utiliza para separar diferentes componentes de una serie temporal. El término **ETS** significa error, tendencia y estacionalidad.

## **¿Qué es la estacionariedad y por qué es importante para el análisis de series temporales?**

La estacionariedad describe el concepto de que la evolución de una serie temporal seguirá siendo la misma en el futuro. En términos matemáticos, una serie temporal es estacionaria cuando sus propiedades estadísticas (media, varianza y covarianza) son independientes del tiempo.

**Una serie *débilmente estacionaria*:**  cumple las siguentes condiciones

1. El valor medio de la serie temporal no depende del tiempo.

2. La covarianza entre dos puntos cualesquiera que estén a la misma distancia es constante. (Por ejemplo, la covarianza entre realizaciones a tres puntos de distancia entre sí debe ser constante independientemente de t).



## **Modelos populares de previsión de series temporales**

- **Naive approach:**  En una previsión ingenua, el valor previsto es simplemente el valor de la observación más reciente. Este método básico suele utilizarse como referencia para evaluar el rendimiento de modelos más sofisticados.

<img src='https://api.wandb.ai/files/iamleonie/images/projects/36935173/477ba4e5.png'>

Fuente: https://wandb.ai/iamleonie/A-Gentle-Introduction-to-Time-Series-Analysis-Forecasting/reports/A-Gentle-Introduction-to-Time-Series-Analysis-Forecasting--VmlldzoyNjkxOTMz


- **Predicción de series temporales basada en la regresión:** Los métodos de previsión de series temporales basados en regresión asumen que la serie temporal tiene una *relación lineal con otras series temporales*. Esta suposición sobre la relación lineal da como resultado un modelo de tendencia pero no de estacionalidad.

- **Media móvil autorregresiva integrada (ARIMA):** ARIMA es una clase de métodos de previsión que modela una serie temporal basándose en sus autocorrelaciones. Los modelos son una combinación de los siguientes componentes: **Autorregresivo (AR):** realiza predicciones basadas en una combinación lineal de sus valores pasados. **Media móvil (MA):** realiza predicciones basadas en una combinación lineal de los errores de predicciones anteriores. **Integrado (I):** realiza predicciones sobre las series temporales diferenciadas, lo que resulta útil cuando una serie temporal no es estacionaria.

- **Aprendizaje automático:** En los últimos años, los modelos de aprendizaje automático han ganado popularidad para la previsión de series temporales. En el concurso Makridakis 2020 (concurso M5), que consiste en una serie de competiciones para evaluar y comparar diferentes métodos de previsión de series temporales, las mejores predicciones fueron producidas por redes neuronales o marcos de Gradient Boosting.

## **Terminologías**

Hay varios términos y conceptos en series de tiempo que debemos conocer. Estos son los siguientes:

1. **Dependencia:** Se refiere a la asociación de dos observaciones de la misma variable en periodos de tiempo anteriores.

2. **Estacionariedad:** Muestra el valor medio de la serie que se mantiene constante a lo largo del periodo de tiempo. Si los efectos pasados se acumulan y los valores aumentan hacia el infinito, entonces no se cumple la estacionariedad.

33. **Time-Series Forecasting:** La previsión de series temporales es el proceso de utilizar un modelo estadístico o de aprendisaje automatico para predecir valores futuros de una serie temporal en función de resultados pasados.

4. **Ajuste de curvas:** la regresión de ajuste de curvas en el análisis de series temporales se utiliza cuando los datos se encuentran en una relación no lineal.

5. **ARIMA:** ARIMA significa Media Móvil Integrada Auto Regresiva.

6. **Diferenciación:** La diferenciación se utiliza para hacer que la serie sea estacionaria y para controlar las autocorrelaciones. Puede haber algunos casos en los análisis de series de tiempo en los que no se requiera la diferenciación y las series sobrediferenciadas pueden producir estimaciones incorrectas.

7. **Especificación:** puede implicar la prueba de las relaciones lineales o no lineales de las variables dependientes mediante el uso de modelos de series temporales como los modelos ARIMA.
