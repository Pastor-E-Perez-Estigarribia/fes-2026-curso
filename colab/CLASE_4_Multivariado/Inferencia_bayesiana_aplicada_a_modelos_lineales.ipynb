
## Introducción a la inferencia bayesiana

---

## 🎯 Objetivo

- Introducir los fundamentos teóricos de la inferencia bayesiana

---

La inferencia bayesiana es un enfoque probabilístico para el análisis de datos que permite actualizar nuestras creencias sobre parámetros desconocidos a partir de la evidencia empírica. A diferencia del paradigma frecuentista, que se basa en estimaciones puntuales y pruebas de hipótesis, el enfoque bayesiano modela explícitamente la incertidumbre mediante distribuciones de probabilidad.

En el contexto de la regresión, este enfoque implica definir distribuciones previas para los coeficientes del modelo y actualizar dichas distribuciones con los datos observados, obteniendo distribuciones posteriores que reflejan tanto el conocimiento previo como la evidencia empírica. Este proceso se formaliza mediante el teorema de Bayes:

$
P(\theta \mid D) = \frac{P(D \mid \theta) \cdot P(\theta)}{P(D)}
$

donde:
- $ \theta $ representa los parámetros del modelo,
- $ D $ son los datos observados,
- $ P(\theta) $ es la distribución previa,
- $ P(D \mid \theta) $ es la verosimilitud,
- $ P(\theta \mid D) $ es la distribución posterior.

La implementación computacional de este enfoque requiere métodos de muestreo como MCMC (Monte Carlo por cadenas de Markov), que permiten aproximar las distribuciones posteriores. Herramientas modernas como Bambi y PyMC facilitan este proceso, permitiendo especificar modelos con sintaxis intuitiva y obtener inferencias robustas, visualizables y reproducibles.

Este práctico se enfoca en aplicar estos conceptos a un conjunto de datos conocido (diamantes), explorando cómo variables como el quilate, el color y la claridad influyen en el precio, y cómo los modelos bayesianos permiten capturar estas relaciones con mayor flexibilidad y transparencia.

## 📘 Formalización de MCMC para regresión lineal múltiple bayesiana

### 1. Modelo general

Consideremos un modelo de regresión lineal múltiple con \( p \) predictores:

$
y_i = \beta_0 + \beta_1 x_{i1} + \beta_2 x_{i2} + \dots + \beta_p x_{ip} + \varepsilon_i
$

donde:
- $ y_i $ es la variable respuesta para la observación $ i $,
- $ x_{ij} $ son los valores de los predictores,
- $ \beta_j $ son los coeficientes del modelo,
- $ \varepsilon_i \sim \mathcal{N}(0, \sigma^2) $ es el error aleatorio.

En notación matricial:

$
\mathbf{y} = \mathbf{X} \boldsymbol{\beta} + \boldsymbol{\varepsilon}
$

---

### 2. Enfoque bayesiano

En el enfoque bayesiano, se especifican distribuciones previas para los parámetros:

- $ \boldsymbol{\beta} \sim \mathcal{N}(\boldsymbol{\mu}_\beta, \boldsymbol{\Sigma}_\beta) $
- $ \sigma^2 \sim \text{Inv-Gamma}(a, b) $ o $ \sigma \sim \text{Half-Cauchy}(0, \tau) $

La verosimilitud del modelo es:

$
P(\mathbf{y} \mid \boldsymbol{\beta}, \sigma^2) = \prod_{i=1}^n \mathcal{N}(y_i \mid \mathbf{x}_i^\top \boldsymbol{\beta}, \sigma^2)
$

La distribución posterior se obtiene mediante el teorema de Bayes:

$
P(\boldsymbol{\beta}, \sigma^2 \mid \mathbf{y}, \mathbf{X}) \propto P(\mathbf{y} \mid \boldsymbol{\beta}, \sigma^2) \cdot P(\boldsymbol{\beta}) \cdot P(\sigma^2)
$

---

### 3. Muestreo MCMC

Dado que la posterior no tiene forma cerrada, se utiliza MCMC para generar muestras de $ (\boldsymbol{\beta}, \sigma^2) $. Los algoritmos comunes incluyen:

- Metropolis-Hastings
- Gibbs Sampling
- Hamiltonian Monte Carlo (HMC), utilizado por PyMC

El procedimiento general es:

1. Inicializar $ \boldsymbol{\beta}^{(0)}, \sigma^{2(0)} $
2. Para $ t = 1 $ hasta $ T $:
   - Proponer nuevos valores $ \boldsymbol{\beta}^{(t)}, \sigma^{2(t)} $
   - Calcular la razón de aceptación:
     $
     r = \frac{P(\mathbf{y} \mid \boldsymbol{\beta}^{(t)}, \sigma^{2(t)}) \cdot P(\boldsymbol{\beta}^{(t)}) \cdot P(\sigma^{2(t)})}{P(\mathbf{y} \mid \boldsymbol{\beta}^{(t-1)}, \sigma^{2(t-1)}) \cdot P(\boldsymbol{\beta}^{(t-1)}) \cdot P(\sigma^{2(t-1)})}
     $
   - Aceptar o rechazar según $ r $

Después del periodo de calentamiento (burn-in), las muestras se consideran provenientes de la distribución posterior.

---

### 4. Inferencia posterior

Con las muestras obtenidas, se pueden calcular:

- Medias posteriores: $ \mathbb{E}[\beta_j] $
- Intervalos de credibilidad: $ P(\beta_j \in [a, b]) $
- Diagnósticos de convergencia: trazas, autocorrelación, R-hat
- Métricas de ajuste: $ R^2 $ bayesiano, LOO, WAIC

## 📐 Verosimilitud en regresión lineal bayesiana con prior gaussiano

### 1. Modelo de regresión lineal

Dado un conjunto de datos $ \{(x_i, y_i)\}_{i=1}^n $, donde cada $ x_i \in \mathbb{R}^p $ representa un vector de predictores y $ y_i \in \mathbb{R}  $ la variable respuesta, el modelo es:

 $
y_i = \beta_0 + \sum_{j=1}^{p} \beta_j x_{ij} + \varepsilon_i, \quad \varepsilon_i \sim \mathcal{N}(0, \sigma^2)
 $

En forma matricial:

 $
\mathbf{y} \sim \mathcal{N}(\mathbf{X} \boldsymbol{\beta}, \sigma^2 \mathbf{I})
 $

donde:
-  $ \mathbf{y} \in \mathbb{R}^n  $ es el vector de respuestas,
-  $ \mathbf{X} \in \mathbb{R}^{n \times p}  $ es la matriz de diseño,
-  $ \boldsymbol{\beta} \in \mathbb{R}^p  $ son los coeficientes,
-  $ \sigma^2  $ es la varianza del error.

---

### 2. Verosimilitud del modelo

La función de verosimilitud para los datos observados  $ \mathbf{y}  $ dado  $ \boldsymbol{\beta}  $ y  $ \sigma^2  $ es:

 $
\mathcal{L}(\boldsymbol{\beta}, \sigma^2 \mid \mathbf{y}, \mathbf{X}) = P(\mathbf{y} \mid \boldsymbol{\beta}, \sigma^2) = \prod_{i=1}^{n} \mathcal{N}(y_i \mid \mathbf{x}_i^\top \boldsymbol{\beta}, \sigma^2)
 $

 $
= \left( \frac{1}{\sqrt{2\pi \sigma^2}} \right)^n \exp\left( -\frac{1}{2\sigma^2} \sum_{i=1}^{n} (y_i - \mathbf{x}_i^\top \boldsymbol{\beta})^2 \right)
 $

---

### 3. Priores gaussianos

Se asume una distribución previa normal para los coeficientes:

 $
\beta_j \sim \mathcal{N}(\mu_j, \tau_j^2), \quad \text{para } j = 0, 1, \dots, p
 $

En forma conjunta:

 $
\boldsymbol{\beta} \sim \mathcal{N}(\boldsymbol{\mu}, \boldsymbol{\Sigma})
 $

donde:
-  $ \boldsymbol{\mu}  $ es el vector de medias previas,
-  $ \boldsymbol{\Sigma} $ es la matriz de covarianzas (diagonal si los priors son independientes).

---

### 4. Posterior bayesiano

$\underbrace{p(\boldsymbol{\theta} \mid \boldsymbol{Y})}_{\text{posterior}} = \frac{\overbrace{p(\boldsymbol{Y} \mid \boldsymbol{\theta})}^{\text{likelihood}}\; \overbrace{p(\boldsymbol{\theta})}^{\text{prior}}}{\underbrace{{p(\boldsymbol{Y})}}_{\text{marginal likelihood}}}$


La distribución posterior se obtiene aplicando el teorema de Bayes:

 $
P(\boldsymbol{\beta}, \sigma^2 \mid \mathbf{y}, \mathbf{X}) \propto \mathcal{L}(\boldsymbol{\beta}, \sigma^2 \mid \mathbf{y}, \mathbf{X}) \cdot P(\boldsymbol{\beta}) \cdot P(\sigma^2)
 $



Esta posterior no tiene forma cerrada en general, por lo que se aproxima mediante métodos MCMC como HMC (Hamiltonian Monte Carlo), especialmente en herramientas como PyMC o Stan.

## Visualización estilo MCMC trace + posterior histogram

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# --- Función posterior para modelo Beta-Binomial ---
def posterior(theta, Y, a=1, b=1):
    if 0 <= theta <= 1:
        prior = stats.beta(a, b).pdf(theta)
        likelihood = stats.bernoulli(theta).pmf(Y).prod()
        return likelihood * prior
    else:
        return -np.inf

# --- Datos simulados: 20 ensayos Bernoulli con p=0.7 ---
np.random.seed(42)
Y = stats.bernoulli(0.7).rvs(20)

# --- Algoritmo Metropolis-Hastings ---
n_iters = 1000
can_sd = 0.05
a = b = 1
theta = 0.5
trace = np.zeros(n_iters)
p_current = posterior(theta, Y, a, b)

for i in range(n_iters):
    theta_candidate = stats.norm(theta, can_sd).rvs()
    p_candidate = posterior(theta_candidate, Y, a, b)
    acceptance_ratio = p_candidate / p_current
    if acceptance_ratio > stats.uniform(0, 1).rvs():
        theta = theta_candidate
        p_current = p_candidate
    trace[i] = theta

# --- Visualización: trazas y histograma ---
fig, axes = plt.subplots(1, 2, figsize=(10, 4), sharey=True)

# Trace plot
axes[0].plot(trace, color='black')
axes[0].set_title("Trace plot")
axes[0].set_ylabel("θ", rotation=0, labelpad=15)
axes[0].set_xlabel("Iteración")

# Histograma rotado
axes[1].hist(trace, bins=30, color='gray', orientation='horizontal', density=True)
axes[1].set_title("Posterior histogram")
axes[1].set_xticks([])
axes[1].set_xlabel("Densidad")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# --- 1. Datos simulados (20 ensayos, p=0.7) ---
np.random.seed(42)
Y = stats.bernoulli(0.7).rvs(20)

# --- 2. Funciones del modelo ---
def posterior(theta, Y, a, b):
    if 0 <= theta <= 1:
        prior = stats.beta(a, b).pdf(theta)
        likelihood = stats.bernoulli(theta).pmf(Y).prod()
        return likelihood * prior
    return -np.inf

def run_mcmc(Y, a, b, n_iters=5000, can_sd=0.05):
    theta = 0.5
    trace = np.zeros(n_iters)
    p_current = posterior(theta, Y, a, b)
    for i in range(n_iters):
        theta_cand = stats.norm(theta, can_sd).rvs()
        p_cand = posterior(theta_cand, Y, a, b)
        # Razón de aceptación
        if p_current == 0:
            ratio = 1.0 if p_cand > 0 else 0.0
        else:
            ratio = p_cand / p_current

        if ratio > stats.uniform(0, 1).rvs():
            theta = theta_cand
            p_current = p_cand
        trace[i] = theta
    return trace

# --- 3. Muestreo con diferentes priors ---
# Prior 1: Uniforme Beta(1, 1)
trace_prior1 = run_mcmc(Y, a=1, b=1)

# Prior 2: Informativo sesgado hacia valores bajos Beta(2, 10)
trace_prior2 = run_mcmc(Y, a=2, b=10)

# --- 4. Visualización comparativa ---
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
x = np.linspace(0, 1, 200)

# Graficar Priors (líneas punteadas)
plt.plot(x, stats.beta(1, 1).pdf(x), 'b--', alpha=0.7, label='Prior 1: Beta(1,1) [Uniforme]')
plt.plot(x, stats.beta(2, 10).pdf(x), 'r--', alpha=0.7, label='Prior 2: Beta(2,10) [Informativo bajo]')

# Graficar Posteriors (KDE con relleno)
sns.kdeplot(trace_prior1, color='blue', fill=True, alpha=0.3, label='Posterior 1 (con Prior 1)')
sns.kdeplot(trace_prior2, color='red', fill=True, alpha=0.3, label='Posterior 2 (con Prior 2)')

plt.title("Efecto del Prior en la Distribución Posterior", fontsize=14)
plt.xlabel("θ (Probabilidad de éxito)")
plt.ylabel("Densidad")
plt.legend()
plt.xlim(0, 1)
plt.show()
```



### 📖 Algoritmo Metropolis-Hastings

El algoritmo Metropolis-Hastings es un método de **Monte Carlo por cadenas de Markov (MCMC)** diseñado para generar muestras de una distribución objetivo $ \pi(\theta) $, incluso cuando esta no puede calcularse de forma explícita.

#### Pasos principales
1. **Inicialización:** se elige un valor inicial $ \theta^{(0)} $.  
2. **Propuesta:** en cada iteración $t$, se genera un candidato $ \theta' $ a partir de una distribución de propuesta $ q(\theta' \mid \theta^{(t)}) $.  
3. **Cálculo de la razón de aceptación:**
   $
   r = \frac{\pi(\theta') \, q(\theta^{(t)} \mid \theta')}{\pi(\theta^{(t)}) \, q(\theta' \mid \theta^{(t)})}
   $
4. **Aceptación o rechazo:**  
   - Se acepta $ \theta' $ con probabilidad $ \min(1, r) $.  
   - Si se acepta, $ \theta^{(t+1)} = \theta' $.  
   - Si se rechaza, $ \theta^{(t+1)} = \theta^{(t)} $.  
5. **Iteración:** se repite el proceso para obtener una cadena de muestras.

#### Propiedades
- La cadena resultante converge a la distribución objetivo $ \pi(\theta) $.  
- Permite aproximar distribuciones posteriores en inferencia bayesiana mediante las muestras generadas.  
- No busca el máximo de la distribución, sino explorar su forma completa.  

---

En síntesis: **Metropolis-Hastings construye una cadena de Markov que, tras suficientes iteraciones, produce muestras distribuidas según la distribución posterior deseada.**




Sea $\pi(\theta)$ la distribución objetivo (por ejemplo, la posterior bayesiana)
y $q(\theta' \mid \theta)$ la distribución de propuesta. El algoritmo se define como:

 1. Inicializar $\theta^{(0)}$.

2.  Para $t = 0,1,\dots,T-1$:

- Generar un candidato $\theta' \sim q(\theta' \mid \theta^{(t)})$.
- Calcular la razón de aceptación:
        $
        r = \frac{\pi(\theta') \, q(\theta^{(t)} \mid \theta')}{\pi(\theta^{(t)}) \, q(\theta' \mid \theta^{(t)})}.
        $
- Aceptar con probabilidad $\alpha = \min(1, r)$:

$\theta^{(t+1)}$=

$\begin{cases}
        \theta', & \text{si } u < \alpha,\quad u \sim \text{Uniform}(0,1), \\
        \theta^{(t)}, & \text{en otro caso}.\end{cases}$


Tras un número suficiente de iteraciones, la cadena $\{\theta^{(t)}\}$
converge en distribución a $\pi(\theta)$.


### Formalización del experimento beta-binomial con muestreo MCMC

#### Modelo y datos

•	Datos observados:

$ Y={y_1,\dots,y_n},\quad y_i\in{0,1},\quad y_i\sim \text{Bernoulli}(\theta)$

•	Parámetro de interés:

$ \theta \in (0,1)$

•	Verosimilitud:

$ p(Y\mid \theta)=\prod_{i=1}n \theta{y_i}(1-\theta)^{1-y_i}$

Priori y posterior

•	Priori Beta(1,1) (uniforme):

$ p(\theta)=\text{Beta}(\theta\mid \alpha,\beta),\quad \alpha=\beta=1$

•	Posterior (no normalizada):

$ p(\theta\mid Y)\propto p(Y\mid \theta),p(\theta) =\theta^{\sum y_i}(1-\theta)^{n-\sum y_i}$

•	Posterior conjugada (forma cerrada):

$ \theta\mid Y \sim \text{Beta}\big(\alpha+\sum y_i,\ \beta+n-\sum y_i\big)$

Inferencia por MCMC (Metropolis-Hastings)

•	Objetivo:

$ \text{aproximar } p(\theta\mid Y) \text{ mediante muestras } {\theta^{(t)}}_{t=1}^T$

•	Propuesta (simétrica):

$ q(\theta'\mid \theta^{(t)})=\mathcal{N}\big(\theta^{(t)},,\sigma^2\big)$

•	Regla de aceptación:

$ a=\min\left(1,\ \frac{p(\theta'\mid Y)}{p(\theta^{(t)}\mid Y)}\right)$

•	Actualización:

## Datos y preprocesado

Usaremos el dataset **penguins** disponible en `seaborn`.  
Centramos la covariable para mejorar la interpretación del intercepto.

**Modelo**

Sea $y_i$ la masa corporal y $x_i$ la longitud del pico centrada. El modelo lineal global es:

$$
\begin{aligned}
y_i &\sim \mathcal{N}(\alpha + \beta x_i,\ \sigma^2)
\end{aligned}
$$

Priori para los parámetros:

\begin{aligned}
\alpha &\sim \mathcal{N}(0,\ \tau_\alpha^2)\\
\beta  &\sim \mathcal{N}(0,\ \tau_\beta^2)\\
\sigma^2 &\sim \text{Inv-Gamma}(a_0, b_0)
\end{aligned}


Usaremos priors débiles: $\tau_\alpha=\tau_\beta=10$, $a_0=1$, $b_0=1$.

---

### Verosimilitud, prior y posterior (expresiones)

**Verosimilitud (likelihood)** para los datos $y=(y_1,\dots,y_N)$ dado $(\alpha,\beta,\sigma^2)$:

$$
p(y \mid \alpha,\beta,\sigma^2)
= \prod_{i=1}^N \mathcal{N}\big(y_i \mid \alpha + \beta x_i,\ \sigma^2\big)
= \\(2\pi\sigma^2)^{-N/2}\exp\!\Big(-\frac{1}{2\sigma^2}\sum_{i=1}^N (y_i - \alpha - \beta x_i)^2\Big).
$$

**Prior conjunta**:

$$
p(\alpha,\beta,\sigma^2)
= \mathcal{N}(\alpha\mid 0,\tau_\alpha^2)\,
  \mathcal{N}(\beta\mid 0,\tau_\beta^2)\,
  \text{Inv-Gamma}(\sigma^2\mid a_0,b_0).
$$

**Posterior (proporcional)**:

$$
p(\alpha,\beta,\sigma^2 \mid y)
\propto p(y \mid \alpha,\beta,\sigma^2)\; p(\alpha,\beta,\sigma^2).
$$

En la práctica numérica conviene trabajar con la **log‑posterior** (evita underflow):

$$
\log p(\alpha,\beta,\sigma^2 \mid y)
= \log p(y \mid \alpha,\beta,\sigma^2) + \log p(\alpha) + \log p(\beta) + \log p(\sigma^2) + C,
$$

donde $C$ es una constante de normalización que no necesitamos para muestrear o comparar valores relativos.

---

A continuación hay una celda de código que implementa funciones en Python para evaluar la verosimilitud, los priors y la log‑posterior (útil para diagnóstico, optimización o muestreo por rejilla).





```python
# Funciones: verosimilitud, prior y log-posterior (usar en Colab)
import numpy as np
from scipy import stats

# Datos (asegúrate de haber ejecutado la celda que define df, X, y, etc.)
# df, X, y ya definidos en celdas previas del notebook

# Prior hyperparameters
tau_alpha = 10.0
tau_beta  = 10.0
a0 = 1.0
b0 = 1.0

# Verosimilitud (log)
def log_likelihood(alpha, beta, sigma2, x, y):
    """
    Log-verosimilitud de los datos bajo el modelo normal.
    alpha, beta: escalares
    sigma2: varianza (escalar, >0)
    x: array de covariables centradas
    y: array de observaciones
    """
    if sigma2 <= 0:
        return -np.inf
    mu = alpha + beta * x
    # suma de cuadrados
    ss = np.sum((y - mu)**2)
    N = len(y)
    ll = -0.5 * N * np.log(2 * np.pi * sigma2) - 0.5 * ss / sigma2
    return float(ll)

# Prior (log)
def log_prior(alpha, beta, sigma2, tau_alpha=tau_alpha, tau_beta=tau_beta, a0=a0, b0=b0):
    """
    Log-prior conjunta: Normal(alpha|0,tau_alpha^2) * Normal(beta|0,tau_beta^2) * Inv-Gamma(sigma2|a0,b0)
    """
    if sigma2 <= 0:
        return -np.inf
    lp_alpha = stats.norm.logpdf(alpha, loc=0.0, scale=tau_alpha)
    lp_beta  = stats.norm.logpdf(beta,  loc=0.0, scale=tau_beta)
    # Inv-Gamma density for sigma2: p(sigma2) = b0^{a0} / Gamma(a0) * sigma2^{-a0-1} * exp(-b0/sigma2)
    # use scipy.stats.invgamma with shape=a0 and scale=b0
    lp_sigma2 = stats.invgamma.logpdf(sigma2, a=a0, scale=b0)
    return float(lp_alpha + lp_beta + lp_sigma2)

# Log-posterior (no normalizado)
def log_posterior(alpha, beta, sigma2, x, y, tau_alpha=tau_alpha, tau_beta=tau_beta, a0=a0, b0=b0):
    """
    Log-posterior no normalizado: log_likelihood + log_prior
    """
    ll = log_likelihood(alpha, beta, sigma2, x, y)
    lp = log_prior(alpha, beta, sigma2, tau_alpha=tau_alpha, tau_beta=tau_beta, a0=a0, b0=b0)
    return ll + lp

# Ejemplo de uso: evaluar log-posterior en un punto (por ejemplo la media posterior aproximada beta_n)
# Asegúrate de ejecutar antes la celda que calcula beta_n y A_inv si quieres usar esos valores.
try:
    alpha0 = float(beta_n[0])   # si beta_n existe (de la conjugada)
    beta0  = float(beta_n[1])
    sigma2_mode = np.median(sigma2_samples) if 'sigma2_samples' in globals() else np.var(y)  # fallback
    print("log-posterior en (alpha0,beta0,sigma2_median):",
          log_posterior(alpha0, beta0, sigma2_mode, df["x"].values, df["mass"].values))
except Exception as e:
    print("Nota: para evaluar el ejemplo asegúrate de haber ejecutado las celdas previas que definen beta_n y sigma2_samples.")
    # no lanzar error, solo informar

# Función auxiliar: evaluar log-posterior en una rejilla (útil para visualizaciones 2D)
def grid_log_posterior(alpha_grid, beta_grid, sigma2_fixed, x, y):
    """
    Devuelve una matriz con log-posterior evaluada en la rejilla (alpha_grid x beta_grid)
    para una sigma2 fija.
    """
    Z = np.empty((len(alpha_grid), len(beta_grid)))
    for i, a in enumerate(alpha_grid):
        for j, b in enumerate(beta_grid):
            Z[i, j] = log_posterior(a, b, sigma2_fixed, x, y)
    return Z

# Ejemplo rápido de rejilla para alpha-beta (usar sigma2 fija = var(y) como aproximación)
# (Descomenta para ejecutar; puede ser costoso si la rejilla es grande)
# alpha_g = np.linspace(alpha0 - 200, alpha0 + 200, 80)
# beta_g  = np.linspace(beta0 - 2, beta0 + 2, 80)
# Z = grid_log_posterior(alpha_g, beta_g, sigma2_mode, df["x"].values, df["mass"].values)
# (luego puedes usar plt.contourf(alpha_g, beta_g, Z.T) para visualizar)
```

## **Posterior conjugada (idea)**

Con priors normales para $(\alpha,\beta)$ y una Inv-Gamma para $\sigma^2$, la posterior marginal de $\sigma^2$ es Inv-Gamma y la condicional $(\alpha,\beta)\mid\sigma^2$ es normal multivariada.  
A continuación implementamos los cálculos estándar para obtener parámetros posteriores, muestrear y construir la posterior predictiva.


```python
# Celda: cálculo de posterior conjugada y muestreo
from scipy import stats

# Datos para el modelo global
x = df["x"].values
y = df["mass"].values
N = len(y)
X = np.column_stack([np.ones(N), x])  # diseño N x 2

# Priors
tau_alpha = 10.0
tau_beta  = 10.0
tau = np.array([tau_alpha, tau_beta])
Tau_inv = np.diag(1.0 / tau**2)  # precisión prior (2x2)

a0 = 1.0
b0 = 1.0

# OLS y cantidades intermedias
y_col = y.reshape(-1,1)
XtX = X.T @ X
XtY = X.T @ y_col
beta_hat = np.linalg.solve(XtX, XtY)  # OLS estimate (2x1)
resid = y_col - X @ beta_hat
SSE = float((resid**2).sum())

# Posterior parameters (conjugada)
A = XtX + Tau_inv
A_inv = np.linalg.inv(A)
beta_n = np.linalg.solve(A, XtY)  # posterior mean (2x1)
a_n = a0 + N/2.0
# b_n formula (incorpora prior)
beta0 = np.zeros((2,1))
b_n = b0 + 0.5*( (y_col.T @ y_col) + (beta0.T @ Tau_inv @ beta0) - (beta_n.T @ A @ beta_n) )
b_n = float(b_n)

print("Posterior Inv-Gamma params for sigma^2: a_n =", a_n, ", b_n =", b_n)
print("Posterior mean for [alpha, beta]:", beta_n.ravel())
```

## **Muestreo posterior y posterior predictiva**

- Muestreamos $\sigma^2$ desde la Inv-Gamma posterior.
- Para cada $\sigma^2$ muestreamos $(\alpha,\beta)$ desde la normal condicional.
- Construimos la distribución predictiva en una rejilla de $x$ y graficamos la media y la banda 95%.


```python
# Celda: muestreo posterior y gráfico predictivo
n_samples = 4000
# Muestrear sigma^2: Inv-Gamma via reciprocals of Gamma
# If X ~ Gamma(shape=a_n, scale=1/b_n) then 1/X ~ Inv-Gamma(a_n, b_n)
gamma_samples = np.random.gamma(shape=a_n, scale=1.0/b_n, size=n_samples)
sigma2_samples = 1.0 / gamma_samples

# Muestrear betas condicionales
beta_samples = np.zeros((n_samples, 2))
for i in range(n_samples):
    s2 = sigma2_samples[i]
    cov = s2 * A_inv
    beta_samples[i,:] = np.random.multivariate_normal(mean=beta_n.ravel(), cov=cov)

# Predicción en rejilla
x_pred = np.linspace(df["x"].min(), df["x"].max(), 80)
X_pred = np.column_stack([np.ones_like(x_pred), x_pred])
y_pred_samples = beta_samples @ X_pred.T  # shape (n_samples, n_x)

y_mean = y_pred_samples.mean(axis=0)
y_low  = np.percentile(y_pred_samples, 2.5, axis=0)
y_high = np.percentile(y_pred_samples, 97.5, axis=0)

# Plot
plt.figure(figsize=(8,5))
plt.scatter(df["x"], df["mass"], alpha=0.5, label="datos")
plt.plot(x_pred, y_mean, color="C1", label="media posterior")
plt.fill_between(x_pred, y_low, y_high, color="C1", alpha=0.25, label="intervalo 95%")
plt.xlabel("bill_length (centrado)")
plt.ylabel("mass (g)")
plt.title("Regresión lineal bayesiana (posterior predictiva)")
plt.legend()
plt.show()
```

## Formalización de la verosimilitud, prior y posterior (y su implementación)

**Modelo**  
Sea $y=(y_1,\dots,y_N)$ y $x=(x_1,\dots,x_N)$ (con $x$ ya centrada). El modelo es
$$
y_i \mid \alpha,\beta,\sigma^2 \sim \mathcal{N}(\alpha + \beta x_i,\ \sigma^2).
$$

---

### Verosimilitud (likelihood)

La verosimilitud conjunta de los datos dado $(\alpha,\beta,\sigma^2)$ es
$$
p(y\mid\alpha,\beta,\sigma^2)
= \prod_{i=1}^N \frac{1}{\sqrt{2\pi\sigma^2}}
\exp\!\Big(-\frac{(y_i - \alpha - \beta x_i)^2}{2\sigma^2}\Big).
$$

La **log‑verosimilitud** (que es la que implementamos por estabilidad numérica) es
$$
\log p(y\mid\alpha,\beta,\sigma^2)
= -\frac{N}{2}\log(2\pi\sigma^2) - \frac{1}{2\sigma^2}\sum_{i=1}^N (y_i - \alpha - \beta x_i)^2.
$$

**Mapeo al código**
- `mu = alpha + beta * x` corresponde a $\alpha + \beta x_i$.  
- `ss = np.sum((y - mu)**2)` corresponde a $\sum_{i=1}^N (y_i - \alpha - \beta x_i)^2$.  
- `ll = -0.5 * N * np.log(2 * np.pi * sigma2) - 0.5 * ss / sigma2` implementa la expresión de la log‑verosimilitud.

---

### Prior conjunta

Usamos priors independientes:
$$
\alpha \sim \mathcal{N}(0,\tau_\alpha^2),\qquad
\beta \sim \mathcal{N}(0,\tau_\beta^2),\qquad
\sigma^2 \sim \text{Inv-Gamma}(a_0,b_0).
$$

En log‑escala:
- Para $\alpha$: $\log p(\alpha)=\log\mathcal{N}(\alpha\mid 0,\tau_\alpha^2)$.  
- Para $\beta$: $\log p(\beta)=\log\mathcal{N}(\beta\mid 0,\tau_\beta^2)$.  
- Para $\sigma^2$: si usamos la parametrización de `scipy.stats.invgamma` con forma $a_0$ y escala $b_0$, la densidad es
$$
p(\sigma^2)=\frac{b_0^{a_0}}{\Gamma(a_0)}(\sigma^2)^{-a_0-1}\exp\!\Big(-\frac{b_0}{\sigma^2}\Big),
$$
y `stats.invgamma.logpdf(sigma2, a=a0, scale=b0)` devuelve $\log p(\sigma^2)$.

**Mapeo al código**
- `lp_alpha = stats.norm.logpdf(alpha, loc=0.0, scale=tau_alpha)`  
- `lp_beta  = stats.norm.logpdf(beta,  loc=0.0, scale=tau_beta)`  
- `lp_sigma2 = stats.invgamma.logpdf(sigma2, a=a0, scale=b0)`  
- `log_prior = lp_alpha + lp_beta + lp_sigma2`

---

### Posterior no normalizada y log‑posterior

La posterior conjunta satisface
$$
p(\alpha,\beta,\sigma^2\mid y)\propto p(y\mid\alpha,\beta,\sigma^2)\,p(\alpha)\,p(\beta)\,p(\sigma^2).
$$

En log‑escala:
$$
\log p(\alpha,\beta,\sigma^2\mid y)
= \log p(y\mid\alpha,\beta,\sigma^2) + \log p(\alpha) + \log p(\beta) + \log p(\sigma^2) + C,
$$
donde $C$ es una constante de normalización que no necesitamos para comparar o muestrear.

**Mapeo al código**
- `log_posterior = log_likelihood(...) + log_prior(...)` implementa la suma de términos anteriores.

---

### Evaluación en rejilla y visualización

Para visualizar la forma de la posterior en $(\alpha,\beta)$ fijamos $\sigma^2=\sigma^2_*$ y evaluamos la log‑posterior en una rejilla:
$$
Z(a,b) = \log p(\alpha=a,\beta=b,\sigma^2=\sigma^2_* \mid y).
$$
La función `grid_log_posterior(alpha_grid, beta_grid, sigma2_fixed, x, y)` calcula exactamente $Z$ punto a punto.

**Consejos numéricos**
- Trabajar en log‑escala evita underflow/overflow.  
- Evitar rejillas demasiado finas (coste computacional $O(n_\alpha n_\beta N)$).  
- Elegir $\sigma^2_*$ como la mediana o la moda aproximada de la posterior de $\sigma^2$ (por ejemplo `np.median(sigma2_samples)`) para obtener contornos representativos.

---

### Uso práctico y diagnóstico

- **Evaluación puntual**: `log_posterior(alpha0,beta0,sigma2_mode, x, y)` permite comparar la plausibilidad relativa de parámetros.  
- **Visualización**: usar `plt.contourf` o `plt.contour` sobre la matriz `Z` para ver los contornos de la log‑posterior en $(\alpha,\beta)$.  
- **Muestreo**: las funciones de log‑posterior son útiles para métodos de optimización (MAP) o para construir propuestas en MCMC personalizado.

# **Regresión bayesiana para la especie Adelie (flipper_length → mass)**

Objetivo: ajustar un modelo lineal bayesiano que predice `mass` a partir de `flipper_length_mm` para la especie *Adelie* usando PyMC.  
El modelo usa `pm.Data` para poder cambiar fácilmente los datos y contiene priors amplios (débilmente informativos).




### Preparación de datos

**Objetivo:** extraer las observaciones para la especie *Adelie*, limpiar valores faltantes y centrar la covariable para mejorar la interpretación del intercepto.

Formalización:
- Observaciones: $y_i$ = masa corporal (g) para $i=1,\dots,N_{\text{Adelie}}$.
- Covariable: $x_i$ = longitud de aleta (flipper length) centrada, $x_i = \text{flipper\_length}_i - \overline{\text{flipper\_length}}$.

Precondiciones: eliminar filas con valores faltantes en las columnas relevantes.


```python
# Celda: preparar datos (Adelie)
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# cargar penguins desde seaborn
penguins = sns.load_dataset("penguins")
# filtrar Adelie y eliminar NA en las columnas relevantes
adelie_mask = (penguins["species"] == "Adelie")
adelie = penguins.loc[adelie_mask, ["flipper_length_mm", "body_mass_g"]].dropna()
adelie = adelie.rename(columns={"flipper_length_mm":"flipper_length", "body_mass_g":"mass"})

# observaciones
adelie_flipper_length_obs = adelie["flipper_length"].values
adelie_mass_obs = adelie["mass"].values

# centrar la covariable (recomendado)
flipper_mean = adelie_flipper_length_obs.mean()
adelie["flipper_c"] = adelie["flipper_length"] - flipper_mean
adelie_flipper_length_obs_c = adelie["flipper_c"].values

print("N Adelie:", len(adelie))
adelie.head()

```

### **Modelo bayesiano (formalización)**

Modelo lineal para *Adelie*:


\begin{aligned}
y_i &\sim \mathcal{N}(\mu_i,\ \sigma^2),\\
\mu_i &= \beta_0 + \beta_1 x_i,\\
\beta_0 &\sim \mathcal{N}(0,\ \tau_{\beta_0}^2),\\
\beta_1 &\sim \mathcal{N}(0,\ \tau_{\beta_1}^2),\\
\sigma &\sim \text{HalfStudentT}(\nu,\ \sigma_{\text{scale}}).
\end{aligned}


En el código usamos $x_i$ centrada para que $\beta_0$ represente la masa esperada en la longitud de aleta promedio. Priors amplios: $\tau_{\beta_0}=\tau_{\beta_1}=4000$, $\nu=100$, $\sigma_{\text{scale}}=2000$.


```python
# Celda: definir y muestrear el modelo con pm.Data (PyMC)
import pymc as pm

with pm.Model() as model_adelie_flipper_regression:
    # usar la versión centrada de la covariable
    adelie_flipper_length = pm.Data("adelie_flipper_length", adelie_flipper_length_obs_c)

    # Priors (amplios / débiles)
    σ = pm.HalfStudentT("σ", nu=100, sigma=2000)        # desviación estándar
    β_0 = pm.Normal("β_0", mu=0, sigma=4000)            # intercepto (g)
    β_1 = pm.Normal("β_1", mu=0, sigma=4000)            # pendiente (g per mm)

    # determinístico para la media
    μ = pm.Deterministic("μ", β_0 + β_1 * adelie_flipper_length)

    # verosimilitud
    mass = pm.Normal("mass", mu=μ, sigma=σ, observed=adelie_mass_obs)

    # muestreo
    inf_data_adelie_flipper_regression = pm.sample(draws=2000, tune=1000, target_accept=0.9, return_inferencedata=True)
```

### **Posterior predictive**

**Propósito:** estimar la distribución de nuevas observaciones $y_{\text{new}}$ dado los datos observados.  
Formalmente, la posterior predictiva para un nuevo $x_*$ es:

$$
p(y_* \mid x_*, y) = \int p(y_* \mid x_*, \beta_0,\beta_1,\sigma)\; p(\beta_0,\beta_1,\sigma \mid y)\; d\beta_0\, d\beta_1\, d\sigma.
$$

En práctica se aproxima por muestreo: para cada muestra $(\beta_0^{(s)},\beta_1^{(s)},\sigma^{(s)})$ se simula $y_*^{(s)} \sim \mathcal{N}(\beta_0^{(s)}+\beta_1^{(s)} x_*,\ \sigma^{(s)2})$ y se resumen las predicciones (media, intervalos 95%).


```python
# Celda: posterior predictive y gráfico de recta con bandas
# extraer muestras de parámetros
β0_s = β0
β1_s = β1
σ_s  = σs
n_samps = len(β0_s)

# rejilla en x centrada
X_pred = np.linspace(adelie["flipper_c"].min(), adelie["flipper_c"].max(), 80)
y_preds = np.zeros((n_samps, X_pred.size))
for i in range(n_samps):
    mu_i = β0_s[i] + β1_s[i] * X_pred
    y_preds[i,:] = np.random.normal(loc=mu_i, scale=σ_s[i])

# resumen predictivo
y_mean = y_preds.mean(axis=0)
y_low  = np.percentile(y_preds, 2.5, axis=0)
y_high = np.percentile(y_preds, 97.5, axis=0)

# Plot: datos y bandas predictivas
plt.figure(figsize=(8,6))
plt.scatter(adelie["flipper_c"], adelie["mass"], alpha=0.6, label="datos Adelie")
plt.plot(X_pred, y_mean, color="C1", label="media posterior")
plt.fill_between(X_pred, y_low, y_high, color="C1", alpha=0.25, label="95% pred. interval")
plt.xlabel("flipper_length (centrado)")
plt.ylabel("mass (g)")
plt.title("Posterior predictive: Adelie flipper → mass")
plt.legend()
plt.show()
```

## Diagnóstico de convergencia: $\hat{R}$ y ESS

**Objetivo:** comprobar que las cadenas MCMC han explorado bien la posterior y que las estimaciones marginales son fiables.

### 1) $\hat{R}$ (Gelman–Rubin potencial de reducción de escala)

- **Definición intuitiva:** $\hat{R}$ compara la variabilidad entre cadenas con la variabilidad dentro de cada cadena. Si todas las cadenas han convergido al mismo objetivo, ambas varianzas deben ser similares y $\hat{R}\approx 1$.
- **Fórmula (versión clásica):** para un parámetro $\theta$ con $m$ cadenas y $n$ muestras por cadena,

  \begin{aligned}
  \bar{\theta}_{\cdot j} &= \frac{1}{n}\sum_{i=1}^n \theta_{ij}\quad(\text{media de la cadena }j)\\
  \bar{\theta}_{\cdot\cdot} &= \frac{1}{m}\sum_{j=1}^m \bar{\theta}_{\cdot j}\\
  B &= \frac{n}{m-1}\sum_{j=1}^m (\bar{\theta}_{\cdot j}-\bar{\theta}_{\cdot\cdot})^2 \quad(\text{varianza entre cadenas})\\
  W &= \frac{1}{m}\sum_{j=1}^m s_j^2 \quad(\text{varianza media dentro de cadenas})\\
  \hat{V} &= \frac{n-1}{n}W + \frac{1}{n}B\\
  \hat{R} &= \sqrt{\frac{\hat{V}}{W}}.
  \end{aligned}
  
- **Interpretación práctica:** valores de $\hat{R}$ cercanos a 1 indican convergencia. Umbrales comunes: $\hat{R}<1.1$ aceptable; idealmente $\hat{R}<1.01$ para inferencias finas.

### 2) ESS (Effective Sample Size)

- **Definición intuitiva:** ESS estima cuántas muestras independientes equivalen a las muestras autocorrelacionadas obtenidas por MCMC. Si las cadenas están muy autocorrelacionadas, ESS será mucho menor que el número bruto de muestras.
- **Fórmula conceptual:** para una cadena combinada de longitud $N$,

  $$\text{ESS} \approx \frac{N}{1 + 2\sum_{t=1}^{T} \rho_t},$$

  donde $\rho_t$ es la autocorrelación lag-$t$ de la cadena y la suma se corta cuando las autocorrelaciones se vuelven insignificantes.
  
- **Interpretación práctica:** ESS grande (p. ej. > 200) es deseable para estimar medias y cuantiles con precisión. Si ESS es pequeño, aumenta `draws` o mejora la mezcla (más `tune`, `target_accept`, reparametrización, centrar/escala).

### 3) Qué mirar en `arviz.summary`
- `r_hat` (o `ess_bulk`, `ess_tail`) en el resumen:  
  - `r_hat` ≈ 1 → convergencia.  
  - `ess_bulk` y `ess_tail` indican ESS para la parte central y las colas de la distribución.
- Si hay parámetros con $\hat{R}$ alto o ESS bajo, investiga trazas, autocorrelación y considera reparametrizar o aumentar `tune/draws`.

A continuación incluimos celdas de código que calculan y muestran $\hat{R}$ y ESS usando `arviz` si está disponible, y con un **fallback** numérico (implementación directa de $\hat{R}$ y una aproximación de ESS por autocorrelación) si `arviz` no está instalado.


```python
# Celda: calcular y mostrar R-hat y ESS (arviz si está, fallback numérico)
import numpy as np

def compute_rhat_from_chains(chains):
    """
    chains: array shape (n_chains, n_draws)
    devuelve rhat escalar para ese parámetro
    """
    m, n = chains.shape
    # medias por cadena
    chain_means = chains.mean(axis=1)
    overall_mean = chain_means.mean()
    # between-chain variance
    B = n * np.sum((chain_means - overall_mean)**2) / (m - 1)
    # within-chain variance
    W = np.sum(np.var(chains, axis=1, ddof=1)) / m
    # marginal posterior variance estimate
    V_hat = ( (n - 1)/n ) * W + (1.0 / n) * B
    R_hat = np.sqrt(V_hat / W)
    return float(R_hat)

def autocorr(x, lag):
    n = len(x)
    x = x - np.mean(x)
    return np.sum(x[:n-lag]*x[lag:]) / np.sum(x*x)

def compute_ess_from_chains(chains, max_lag=100):
    """
    chains: array shape (n_chains, n_draws)
    aproximación ESS combinando cadenas y sumando autocorrelaciones
    """
    m, n = chains.shape
    # combinar cadenas en una sola secuencia intercalada para reducir sesgo
    combined = chains.reshape(-1, order='F')  # interleave chains
    # calcular autocorrelaciones hasta max_lag o hasta que se vuelvan negativas consecutivas
    rho_sum = 0.0
    for lag in range(1, min(max_lag, n-1)):
        rho = autocorr(combined, lag)
        if rho <= 0:
            break
        rho_sum += rho
    N_eff = len(combined) / (1 + 2 * rho_sum)
    return float(N_eff)

# Función principal: resumen diagnóstico para un InferenceData o diccionario de muestras
def summary_diagnostics(idata, var_names=None):
    """
    idata: si es arviz.InferenceData se usa arviz; si es dict con posterior arrays, se usa fallback.
    var_names: lista de variables a evaluar (ej: ["β_0","β_1","σ"])
    """
    try:
        import arviz as az
        # usar arviz summary (incluye r_hat y ess)
        if var_names is None:
            print("Resumen (arviz):")
            print(az.summary(idata))
        else:
            print("Resumen (arviz) para variables:", var_names)
            print(az.summary(idata, var_names=var_names))
    except Exception:
        print("arviz no disponible: usando cálculo numérico aproximado de R-hat y ESS.")
        # intentar extraer posterior como dict: posterior[var].values shape (chain, draw, *shape)
        posterior = None
        try:
            posterior = idata.posterior
        except Exception:
            # si idata es dict de arrays
            posterior = idata
        for var in (var_names or posterior.keys()):
            try:
                arr = posterior[var].values  # shape (chain, draw, ...)
                # si tiene dimensiones extra, aplanar la dimensión de parámetros (ej: vector)
                if arr.ndim == 3:
                    # (chain, draw, param_dim) -> iterar sobre param_dim
                    nparam = arr.shape[2]
                    for k in range(nparam):
                        chains = arr[:,:,k]
                        rhat = compute_rhat_from_chains(chains)
                        ess = compute_ess_from_chains(chains)
                        print(f"{var}[{k}]: R-hat ≈ {rhat:.3f}, ESS ≈ {ess:.1f}")
                elif arr.ndim == 2:
                    chains = arr
                    rhat = compute_rhat_from_chains(chains)
                    ess = compute_ess_from_chains(chains)
                    print(f"{var}: R-hat ≈ {rhat:.3f}, ESS ≈ {ess:.1f}")
                else:
                    print(f"{var}: formato inesperado arr.ndim={arr.ndim}")
            except Exception as e:
                print(f"No se pudo procesar variable {var}: {e}")

# Ejemplo de uso (si tienes un InferenceData llamado inf_data_adelie_flipper_regression)
# summary_diagnostics(inf_data_adelie_flipper_regression, var_names=["β_0","β_1","σ"])
```

# **Ajustar automáticamente el mismo modelo (estructura y priors) por cada especie del dataset**

## Objetivo

Ajustar automáticamente el mismo modelo (estructura y priors) por cada especie del dataset `penguins`, extraer las muestras de la pendiente $\beta_1$ para cada especie y producir un **forest plot** comparativo de las medias posteriores y sus intervalos creíbles 95%.

**Flujo**
1. Iterar sobre especies únicas en `penguins`.
2. Para cada especie:
   - filtrar datos, centrar la covariable `flipper_length`,
   - definir el modelo PyMC (misma estructura que para Adelie),
   - muestrear y guardar las muestras de `β_1`.
3. Construir un `forest plot` con la media posterior y el 95% HDI de `β_1` para cada especie.


```python
# Celda: ajustar modelos por especie y recolectar muestras de beta_1
import pymc as pm
import arviz as az
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# cargar datos (si no están en memoria)
penguins = sns.load_dataset("penguins")
penguins = penguins.dropna(subset=["flipper_length_mm","body_mass_g","species"])
penguins = penguins.rename(columns={"flipper_length_mm":"flipper_length","body_mass_g":"mass"})

species_list = penguins["species"].unique().tolist()
results = {}  # guardará dict por especie: {'beta1_samples': array, 'idata': InferenceData}

for sp in species_list:
    sub = penguins[penguins["species"] == sp].copy()
    # centrar covariable
    sub["flipper_c"] = sub["flipper_length"] - sub["flipper_length"].mean()
    x_obs = sub["flipper_c"].values
    y_obs = sub["mass"].values
    print(f"Ajustando modelo para especie: {sp} (N={len(sub)})")
    with pm.Model() as model_sp:
        flipper = pm.Data("flipper", x_obs)
        σ = pm.HalfStudentT("σ", nu=100, sigma=2000)
        β_0 = pm.Normal("β_0", mu=0, sigma=4000)
        β_1 = pm.Normal("β_1", mu=0, sigma=4000)
        μ = pm.Deterministic("μ", β_0 + β_1 * flipper)
        mass = pm.Normal("mass", mu=μ, sigma=σ, observed=y_obs)
        # muestreo (ajusta draws/tune si es necesario)
        idata_sp = pm.sample(draws=1500, tune=1000, target_accept=0.9, return_inferencedata=True)
    # extraer muestras de β_1
    beta1 = idata_sp.posterior["β_1"].stack(sample=("chain","draw")).values
    results[sp] = {"beta1_samples": beta1, "idata": idata_sp, "n": len(sub)}

print("Ajustes completados para todas las especies.")
```

```python
# Celda: forest plot comparativo de pendientes (beta_1)
def hdi_from_samples(samples, prob=0.95):
    lo = np.percentile(samples, (1-prob)/2*100)
    hi = np.percentile(samples, (1+(prob))/2*100)
    return lo, hi

# preparar datos para el plot
rows = []
for sp, info in results.items():
    s = info["beta1_samples"]
    mean = s.mean()
    lo, hi = hdi_from_samples(s, 0.95)
    rows.append({"species": sp, "mean": mean, "lo": lo, "hi": hi, "n": info["n"]})

df_forest = pd.DataFrame(rows).sort_values("mean")
y_pos = np.arange(len(df_forest))

# plot
plt.figure(figsize=(8, 4 + 0.6*len(df_forest)))
plt.hlines(y=y_pos, xmin=df_forest["lo"], xmax=df_forest["hi"], color="C0", alpha=0.8)
plt.plot(df_forest["mean"], y_pos, "o", color="C1")
plt.yticks(y_pos, df_forest["species"])
plt.xlabel("β_1 (pendiente) — media posterior y 95% HDI")
plt.title("Forest plot comparativo de pendientes por especie")
plt.gca().invert_yaxis()
# anotar N por especie
for i, (_, row) in enumerate(df_forest.iterrows()):
    plt.text(row["hi"] + 0.02*abs(row["hi"]), i, f"N={int(row['n'])}", va="center")
plt.tight_layout()
plt.show()
```

```python
# Celda: guardar muestras de beta1 por especie (.npz y .csv)
import numpy as np
import pandas as pd
import os

# Nombre de archivos de salida
npz_path = "beta1_samples_by_species.npz"
csv_path = "beta1_samples_long.csv"

# Si no existe 'results', intentar recomputar (cuidado: puede tardar)
if 'results' not in globals():
    print("No se encontró 'results' en memoria. Recalculando modelos por especie (esto puede tardar)...")
    # Código mínimo para recalcular results (misma estructura que en el notebook anterior)
    import pymc as pm, seaborn as sns
    penguins = sns.load_dataset("penguins").dropna(subset=["flipper_length_mm","body_mass_g","species"])
    penguins = penguins.rename(columns={"flipper_length_mm":"flipper_length","body_mass_g":"mass"})
    species_list = penguins["species"].unique().tolist()
    results = {}
    for sp in species_list:
        sub = penguins[penguins["species"] == sp].copy()
        sub["flipper_c"] = sub["flipper_length"] - sub["flipper_length"].mean()
        x_obs = sub["flipper_c"].values
        y_obs = sub["mass"].values
        with pm.Model() as model_sp:
            flipper = pm.Data("flipper", x_obs)
            σ = pm.HalfStudentT("σ", nu=100, sigma=2000)
            β_0 = pm.Normal("β_0", mu=0, sigma=4000)
            β_1 = pm.Normal("β_1", mu=0, sigma=4000)
            μ = pm.Deterministic("μ", β_0 + β_1 * flipper)
            mass = pm.Normal("mass", mu=μ, sigma=σ, observed=y_obs)
            idata_sp = pm.sample(draws=1500, tune=1000, target_accept=0.9, return_inferencedata=True)
        beta1 = idata_sp.posterior["β_1"].stack(sample=("chain","draw")).values
        results[sp] = {"beta1_samples": beta1, "idata": idata_sp, "n": len(sub)}
    print("Reajuste completado.")

# Construir dict para .npz y DataFrame long para .csv
npz_dict = {}
rows = []
for sp, info in results.items():
    arr = np.asarray(info["beta1_samples"]).ravel()
    npz_dict[sp] = arr
    rows.append(pd.DataFrame({"species": sp, "beta1": arr}))

# Guardar .npz
np.savez_compressed(npz_path, **npz_dict)
print(f"Guardado: {npz_path} (arrays por especie)")

# Concatenar y guardar CSV (long format)
df_long = pd.concat(rows, ignore_index=True)
df_long.to_csv(csv_path, index=False)
print(f"Guardado: {csv_path} (formato long: species, beta1)")

# Mostrar resumen rápido
print("\nResumen de muestras guardadas (por especie):")
for sp, arr in npz_dict.items():
    print(f" - {sp}: {arr.size} muestras, mean={arr.mean():.3f}, 95% HDI≈[{np.percentile(arr,2.5):.3f}, {np.percentile(arr,97.5):.3f}]")

# Listar archivos en el directorio actual (Colab)
print("\nArchivos en el directorio actual:")
print("\n".join(sorted(os.listdir("."))[:50]))
```

# **Regresión lineal bayesiana con PyStan**

**Objetivos**

- Reproducir el ejemplo de regresión lineal del notebook original en PyStan.
- Entender la formulación Bayesiana, generar datos sintéticos, compilar el modelo Stan, muestrear con NUTS y diagnosticar la convergencia.
- Visualizar la incertidumbre en la recta de regresión y en los parámetros.


## Formalización del modelo

Modelo lineal simple para observaciones $(x_i,y_i)$:

\begin{aligned}
y_i &\sim \mathcal{N}(\alpha + \beta x_i,\ \sigma^2)\\
\alpha &\in \mathbb{R},\quad \beta \in \mathbb{R},\quad \sigma>0.
\end{aligned}


Queremos inferir la posterior $p(\alpha,\beta,\sigma\mid y,x)$ usando MCMC (NUTS).


### Instalación (ejecutar solo si `pystan` no está disponible)

En Colab puede que necesites instalar `pystan`. Ejecuta la celda siguiente solo si `import pystan` falla.
Después de instalar reinicia el runtime (Runtime → Restart runtime) y vuelve a ejecutar las celdas.


## **Generación de datos sintéticos**

Elegimos valores "verdaderos" $(\alpha^*,\beta^*,\sigma^*)$ y generamos:
$$
x_i \sim \text{Uniform}(0,10),\qquad y_i = \alpha^* + \beta^* x_i + \varepsilon_i,\quad \varepsilon_i\sim\mathcal{N}(0,\sigma^{*2}).
$$

Esto nos permite comprobar si el muestreador recupera los parámetros verdaderos.


```python
import stan
import numpy as np
import pandas as pd
import nest_asyncio

# Permite ejecutar PyStan 3 dentro de un entorno Jupyter sin problemas de event loops
nest_asyncio.apply()

# 1. Generar datos sintéticos (Regresión lineal simple)
np.random.seed(42)
N = 100
x = np.random.uniform(0, 10, size=N)
alpha_real = 2.5
beta_real = 1.2
sigma_real = 0.8

# y = alpha + beta * x + ruido
y = alpha_real + beta_real * x + np.random.normal(0, sigma_real, size=N)

# Plot
plt.figure(figsize=(7,5))
plt.scatter(x, y, alpha=0.7)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Datos sintéticos: scatter")
plt.show()
```

## **Modelo Stan (estructura)**

En Stan definimos bloques `data`, `parameters` y `model`.  
- `data` declara las entradas (N, x, y).  
- `parameters` declara $\alpha,\beta,\sigma$ (con $\sigma>0$).  
- `model` contiene la verosimilitud $y \sim \text{normal}(\alpha + \beta x, \sigma)$ y opcionalmente priors.

A continuación el código Stan como cadena de texto.


```python

# Preparar el diccionario de datos para Stan
stan_data = {
    'N': N,
    'x': x.tolist(),
    'y': y.tolist()
}

# 2. Definir el modelo Stan
model_code = """
data {
  int<lower=0> N;
  vector[N] x;
  vector[N] y;
}
parameters {
  real alpha;
  real beta;
  real<lower=0> sigma;
}
model {
  // Priors débiles
  alpha ~ normal(0, 10);
  beta ~ normal(0, 10);
  sigma ~ cauchy(0, 5);

  // Verosimilitud (Likelihood)
  y ~ normal(alpha + beta * x, sigma);
}
"""
```

## Compilar y muestrear con PyStan

- Convertimos el modelo Stan a un objeto compilado (`StanModel`) y llamamos a `sampling`.
- Parámetros importantes: `iter` (total por cadena), `chains`, `warmup` (burn-in), `thin`, `seed`.
- Tras el muestreo extraemos `n_eff` (ESS) y `Rhat` para diagnosticar convergencia.


```python

# 3. Compilar el modelo usando la sintaxis de PyStan 3
print("Compilando el modelo con stan.build()...")
posterior = stan.build(model_code, data=stan_data, random_seed=42)

# 4. Muestrear de la posterior
print("Generando muestras con posterior.sample()...")
fit = posterior.sample(num_chains=4, num_samples=1000)

# 5. Extraer y explorar los resultados
# En PyStan 3 podemos convertir directamente el objeto 'fit' a un DataFrame de Pandas
df_fit = fit.to_frame()

print("\n--- Parámetros Reales ---")
print(f"Alpha: {alpha_real}, Beta: {beta_real}, Sigma: {sigma_real}")

print("\n--- Resumen de los Parámetros Estimados ---")
# Mostramos las estadísticas básicas de los parámetros que nos interesan
print(df_fit[['alpha', 'beta', 'sigma']].describe().T[['mean', 'std', '25%', '50%', '75%']])
```

## Interpretación del resumen

- `n_eff` (o `n_eff` en la tabla): estimación del tamaño efectivo de muestra (ESS). Valores altos indican muestras efectivas suficientes.  
- `Rhat` (o `Rhat` en la tabla): estadístico de Gelman–Rubin; valores cercanos a 1 indican convergencia entre cadenas.  
Si alguna variable muestra `Rhat` significativamente > 1.1 o `n_eff` muy bajo, conviene aumentar `iter`/`warmup` o revisar parametrización.


```python
# Extraer solo las muestras del parámetro alpha
alpha_samples_stan = df_fit['alpha']

# Mostrar las primeras muestras y un resumen estadístico
print("Primeras muestras de alpha:")
print(alpha_samples_stan.head())

print("\nResumen estadístico de alpha:")
print(alpha_samples_stan.describe())
```

```python
# Celda: función para traza + posterior y uso para alpha, beta, sigma
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

def plot_trace(param, name='parameter'):
    mean = np.mean(param)
    median = np.median(param)
    cred_min, cred_max = np.percentile(param, 2.5), np.percentile(param, 97.5)
    plt.figure(figsize=(10,4))
    plt.subplot(1,2,1)
    plt.plot(param, alpha=0.6)
    plt.xlabel('samples'); plt.ylabel(name)
    plt.axhline(mean, color='r', lw=2, linestyle='--', label='mean')
    plt.axhline(median, color='c', lw=2, linestyle='--', label='median')
    plt.axhline(cred_min, color='k', linestyle=':', alpha=0.6)
    plt.axhline(cred_max, color='k', linestyle=':', alpha=0.6)
    plt.legend()
    plt.title(f'Trace: {name}')
    plt.subplot(1,2,2)
    plt.hist(param, bins=30, density=True, alpha=0.6)
    sns.kdeplot(param, fill=True)
    plt.axvline(mean, color='r', lw=2, linestyle='--')
    plt.axvline(median, color='c', lw=2, linestyle='--')
    plt.title(f'Posterior: {name}')
    plt.tight_layout()
    plt.show()

# Usar la función
plot_trace(df_fit['alpha'], 'alpha')
plot_trace(df_fit['beta'], 'beta')  # Se usa solo la columna de beta
plot_trace(df_fit['sigma'], 'sigma')
```

### Comparación con Scikit-Learn (Frecuentista)

Para comprobar cómo se alinean las estimaciones de nuestro modelo Bayesiano con un enfoque clásico, ajustaremos un modelo de mínimos cuadrados ordinarios (OLS) usando `sklearn.linear_model.LinearRegression`.

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
import numpy as np

# sklearn requiere que la matriz de características X sea 2D (n_samples, n_features)
x_sk = x.reshape(-1, 1)

# Inicializamos y ajustamos el modelo frecuentista
modelo_sk = LinearRegression()
modelo_sk.fit(x_sk, y)

# Extraemos los parámetros
alpha_sk = modelo_sk.intercept_
beta_sk = modelo_sk.coef_[0]

# Calculamos sigma (desviación estándar de los residuos)
residuos = y - modelo_sk.predict(x_sk)
sigma_sk = np.std(residuos, ddof=2)  # ddof=2 porque estimamos 2 parámetros (alpha y beta)

print("--- Resultados Scikit-Learn (Frecuentista) ---")
print(f"Alpha (Intercepto): {alpha_sk:.4f}")
print(f"Beta (Pendiente)  : {beta_sk:.4f}")
print(f"Sigma (Ruido)     : {sigma_sk:.4f}")

print("\n--- Resultados PyStan (Bayesiano, Media Posterior) ---")
print(f"Alpha (Intercepto): {df_fit['alpha'].mean():.4f}")
print(f"Beta (Pendiente)  : {df_fit['beta'].mean():.4f}")
print(f"Sigma (Ruido)     : {df_fit['sigma'].mean():.4f}")

print("\n--- Parámetros Verdaderos ---")
print(f"Alpha : {alpha_real:.4f}")
print(f"Beta  : {beta_real:.4f}")
print(f"Sigma : {sigma_real:.4f}")
```

**Interpretación:**
Como puedes observar, las estimaciones puntuales (medias) de MCMC convergen prácticamente a los mismos valores que el ajuste por Mínimos Cuadrados Ordinarios (OLS) de `sklearn`. La gran ventaja del modelo Bayesiano (Stan) es que nos proporciona toda la **distribución posterior**, permitiéndonos cuantificar de manera natural la incertidumbre de estas estimaciones sin recurrir a supuestos asintóticos, e integrando nuestros priors si así lo deseamos.

### Visualización de la incertidumbre de la recta de regresión

Usando las muestras posteriores de $\alpha$ y $\beta$, podemos trazar la recta de regresión media y un intervalo de credibilidad del 95% (HPD) para observar cómo la incertidumbre de los parámetros se traduce en incertidumbre sobre la recta ajustada.

```python
import matplotlib.pyplot as plt
import numpy as np

# Extraer las muestras de la posterior del DataFrame de PyStan
alphas = df_fit['alpha'].values
betas = df_fit['beta'].values

# Crear una rejilla de valores 'x' para generar las predicciones de la recta
x_pred = np.linspace(x.min(), x.max(), 100)

# Calcular las rectas de regresión (mu = alpha + beta * x) para cada muestra
# Usamos broadcasting: alphas y betas tienen forma (N_samples,), x_pred (100,)
# mu_pred tendrá forma (N_samples, 100)
mu_pred = alphas[:, None] + betas[:, None] * x_pred[None, :]

# Calcular la media y los percentiles 2.5% y 97.5% (Intervalo 95%) para cada punto x
mu_mean = mu_pred.mean(axis=0)
mu_low = np.percentile(mu_pred, 2.5, axis=0)
mu_high = np.percentile(mu_pred, 97.5, axis=0)

# Crear la visualización
plt.figure(figsize=(8, 6))
plt.scatter(x, y, alpha=0.5, label='Datos sintéticos')

# Graficar la banda de incertidumbre del 95%
plt.fill_between(x_pred, mu_low, mu_high, color='C1', alpha=0.3, label='Incertidumbre 95% de la media')

# Graficar la recta de regresión promedio
plt.plot(x_pred, mu_mean, color='C1', linewidth=2, label='Recta de regresión (media posterior)')

plt.xlabel('x')
plt.ylabel('y')
plt.title('Regresión Lineal Bayesiana (PyStan): Incertidumbre de la recta')
plt.legend()
plt.grid(alpha=0.2)
plt.show()
```

## 📊 Comparación entre modelado frecuentista y bayesiano por MCMC

| Aspecto                          | Enfoque Frecuentista                          | Enfoque Bayesiano por MCMC                          |
|----------------------------------|-----------------------------------------------|-----------------------------------------------------|
| **Fundamento teórico**           | Basado en estimación puntual y pruebas de hipótesis | Basado en actualización de creencias mediante el teorema de Bayes |
| **Estimación de parámetros**     | Punto óptimo (máxima verosimilitud, mínimos cuadrados) | Distribución posterior de parámetros |
| **Incertidumbre**                | Intervalos de confianza (basados en asintóticos) | Intervalos de credibilidad (basados en la posterior) |
| **Interpretación de coeficientes** | “Si repitiéramos el experimento…”             | “Dado lo que sabemos y observamos…”                 |
| **Priores**                      | No se utilizan                                | Se especifican explícitamente (informativos o no informativos) |
| **Verosimilitud**                | Maximizada directamente                       | Combinada con el prior para formar la posterior     |
| **Inferencia**                   | Basada en valores críticos y p-valores        | Basada en densidades posteriores y probabilidades directas |
| **Computación**                  | Rápida y directa (algoritmos cerrados)        | Intensiva (requiere muestreo MCMC)                  |
| **Diagnóstico**                  | Supuestos clásicos (normalidad, homocedasticidad) | Trazas, autocorrelación, R-hat, divergencias        |
| **Flexibilidad**                 | Limitada para modelos complejos o jerárquicos | Alta: permite modelos mixtos, no lineales, jerárquicos |
| **Software típico**              | `lm()`, `glm()`, `lme4`, `statsmodels`, `sklearn` | `brms`, `rstan`, `PyMC`, `Bambi`, `Stan`            |
| **Resultado final**              | Estimación puntual con error estándar         | Distribución completa de cada parámetro             |
