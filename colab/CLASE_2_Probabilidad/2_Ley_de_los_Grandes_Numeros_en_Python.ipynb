## 📘 Introducción breve

La **Ley de los Grandes Números (LGN)** establece que, al repetir un experimento aleatorio muchas veces, la frecuencia relativa de un evento tiende a estabilizarse cerca de su probabilidad teórica.

### 📐 Formulación

Si $ X_1, X_2, ..., X_n $ son variables aleatorias independientes con esperanza $ \mu $, entonces:

$
\lim_{n \to \infty} \frac{1}{n} \sum_{i=1}^{n} X_i = \mu
$

## 🧪 Simulación 1: Lanzamiento de moneda

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Espacio muestral
moneda = ["cara", "cruz"]

# Simulación de 10,000 lanzamientos
n = 10000
acumular = np.random.choice(moneda, size=n, replace=True)

# Visualización de convergencia
resultados = []
for i in range(1, 40):
    muestra = np.random.choice(moneda, size=i, replace=True)
    proporciones = np.mean(muestra == "cara")
    resultados.append(proporciones)

# Gráfico de convergencia
plt.plot(resultados, label="Proporción de 'cara'")
plt.axhline(y=0.5, color='red', linestyle='--', label="Probabilidad teórica")
plt.xlabel("Número de lanzamientos")
plt.ylabel("Proporción")
plt.title("Ley de los Grandes Números: Moneda")
plt.legend()
plt.grid(True)
plt.show()
```

## 📊 Simulación 2: Distribución de proporciones

```python
plt.figure()
plt.hist(resultados, bins=15, density=True, alpha=0.6, color='skyblue')
plt.axvline(x=0.5, color='red', linestyle='--')
plt.title("Distribución de proporciones (moneda)")
plt.xlabel("Proporción de 'cara'")
plt.ylabel("Densidad")
plt.grid(True)
plt.show()
```

### Formalización matemática breve

- Sea $X_1, X_2$, … i.i.d. con $P(X_i = 1) = p$ y $P(X_i = 0) = 1−p$ (moneda con cara=1, cruz=0, p=0.5).  
- Media muestral para un tamaño n: $\overline{X}_n = \frac{1}{n}\sum_{i=1}^n X_i.$  
- Ley de los Grandes Números (forma débil): para todo ε > 0,
  $
  \Pr\big(|\overline{X}_n - p| > \varepsilon\big) \to 0 \quad\text{cuando } n\to\infty.
  $
- Convergencia en probabilidad significa que la proporción observada tiende a p en la mayoría de las realizaciones conforme n crece; sin embargo para muestras finitas habrá variabilidad entre las N trayectorias.

### 📐 Formulación matemática

$
