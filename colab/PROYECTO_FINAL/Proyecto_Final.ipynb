### 🎯 Objetivo general
Proponer una hipótesis de investigación en un estudio hipotético (simulado) o con datos reales y evaluar computacionalmente.

---

### 📋 Instrucciones

1. **Diseño de un estudio hipotético**
   - Proponga un escenario donde se evalúe el impacto de una técnica, modelo o intervención en un sistema.

   - Ejemplos posibles:
     - Comparación entre dos algoritmos de clasificación.
     - Evaluación del efecto de una técnica de preprocesamiento sobre la precisión de un modelo.
     - Diseños factoriales o multifactorales
     - Por ajuste y estimación de parametos (métodos de minimos cuadrados, maxima verosimilitus o inferencias bayesiana)

2. **Formulación de la pregunta de investigación**
   - Redacte una pregunta clara que relacione una variable categórica (modelo, técnica, grupo) con una variable de resultado.

3. **Planteamiento de hipótesis de investigación y predicción**
   - Hipótesis de investigación: enuncie una afirmación que responda a la pregunta.
   - Predicción: indique qué patron espera observar en los resultados.

4. **Formalización de hipótesis estadísticas**
   - Hipótesis nula ($ H_0 $): no hay diferencia entre condiciones.
   - Hipótesis alternativa ($ H_1 $): existe una diferencia significativa.
   - Defina si la prueba será unilateral o bilateral.

   Obs.: la hipotesis estadistica es la formalización de la predicción.

5. **Simulación de datos o importación de datos reales**
   - Genere grupos de datos numéricos representando el resultado de cada condición, o
   - Importe sus datos
   - Puede usar distribuciones normales, lognormales, o simulaciones basadas en resultados de modelos de IA.

6. **Aplicación computacional de la prueba**
   - Calcule la estadística de prueba (ej. diferencia de medias, precisión, etc.).
   - Estime el valor p como por ejemplo la proporción de permutaciones con estadística igual o más extrema que la observada.


7. **Interpretación de resultados**
   - Compare el valor p con el nivel de significancia ($ \alpha = 0.05 $).
   - Decida si se rechaza o no la hipótesis nula.
   - Relacione el resultado con su predicción inicial.

8. **Reflexión técnica**
   - ¿El resultado apoya su hipótesis de investigación?
   - ¿Qué implicancias tendría en un contexto de toma de decisiones?
   - ¿Qué limitaciones tiene sus resultados y cómo podría mejorarla?
